```python
import zipfile
import rarfile
def get_main_folders(zip_file_path):
    main_folders = set()
    csv_files = set()
    if zip_file_path.lower().endswith('.rar'):
        with rarfile.RarFile(zip_file_path, 'r') as rar_file:
            print('This is rar file')
            for name in rar_file.namelist():
                if '/' in name:
                    main_folder = name.split('/')[1]
                    main_folders.add(main_folder)
                if '/' not in name:
                    csv_files.add(name)
    else:
        with zipfile.ZipFile(zip_file_path, 'r') as zip_file:
            print('This is zipfile')
            for name in zip_file.namelist():
                if '/' in name:
                    main_folder = name.split('/')[1]
                    # print(os.listdir(main_folder))
                    # print('THis is zipfile')
                    main_folders.add(main_folder)
                if '/' not in name:
                    csv_files.add(name)
    
    return main_folders, csv_files

zip_file_path = 'Sample.zip'
main_folders, csv_files = get_main_folders(zip_file_path)

print("Main folders inside the ZIP file:")
for folder in main_folders:
    pass
    # print(folder)
print(csv_files)
```

```python
import zipfile
import rarfile

def find_image_directories(file_path):
    image_directories = set()
    if file_path.lower().endswith('.rar'):
        with rarfile.RarFile(file_path, 'r') as rar_file:
            for name in rar_file.namelist():
                if name.lower().endswith(('.jpg', '.jpeg', '.png', '.gif')):
                    parts = name.split('/')
                    if len(parts) > 1:
                        directory = '/'.join(parts[:-1])
                        last_directory = directory.split('/')
                        direc = last_directory[-1]
                        image_directories.add(direc)
    else: 
        with zipfile.ZipFile(file_path, 'r') as zip_file:
            for name in zip_file.namelist():
                if name.lower().endswith(('.jpg', '.jpeg', '.png', '.gif')):
                    parts = name.split('/')
                    if len(parts) > 1:
                        directory = '/'.join(parts[:-1])
                        last_directory = directory.split('/')
                        direc = last_directory[-1]
                        image_directories.add(direc)

    
    # else:
    #     print('File format error')
    return image_directories

# file_path = ''  # Replace with the actual path to your ZIP or RAR file
image_directories = find_image_directories(zip_file_path)

print("Directories to the images in the ZIP/RAR file:")
classes = []
for directory in image_directories:
    classes.append(directory)
print(classes)
```
